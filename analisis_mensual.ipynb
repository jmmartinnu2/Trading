REGRESION LINEAL PARA PREDECIR PRECIO DE CIERRE CON VIX_CLOSE A LA BAJA

```python
#librerias para regresion linea
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

#un primer modelo de regresion linea para predecir precio de cierre
x= df['close'].values.reshape(-1,1)
y = df['vwap'].values.reshape(-1,1)

sc_x = StandardScaler()
sc_y = StandardScaler()

x_std = sc_x.fit_transform(x)
y_std = sc_y.fit_transform(y)

slr = LinearRegression()
slr.fit(x_std, y_std)


num_vix = 29.70
num_vix_std = sc_x.transform(np.array([num_vix]).reshape(-1,1))
```

```python
plt.scatter(x_std,y_std)
plt.plot(x_std,slr.predict(x_std), color='b')
```

```python
vwap_stat = 3672.25
vwap_stat_std = sc_x.transform(np.array([vwap_stat]).reshape(-1,1))
print('el precio de cierre mensaul se situara en: ', sc_y.inverse_transform(slr.predict(vwap_stat_std)))
```
